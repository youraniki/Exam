```python
reviews = list(data["review"])
banks = list(data["bank"])
values = list(data["rating_value"])
banks_n_reviews = dict()
for i in range(len(banks)):
    if banks[i] not in banks_n_reviews:
        banks_n_reviews[banks[i]] = {"min_value": values[i],
                                "max_value": values[i],
                                "min_review": reviews[i],
                                "max_review": reviews[i]}
    else:
        if values[i] < banks_n_reviews[banks[i]]["min_value"]:
            banks_n_reviews[banks[i]]["min_value"] = values[i]
            banks_n_reviews[banks[i]]["min_review"] = reviews[i]
        elif values[i] > banks_n_reviews[banks[i]]["max_value"]:
            banks_n_reviews[banks[i]]["max_value"] = values[i]
            banks_n_reviews[banks[i]]["max_review"] = reviews[i]
```

```python
reviews_for_file = []
values_for_file = []
for bank in banks_n_reviews:
    if banks_n_reviews[bank]["min_review"] == banks_n_reviews[bank]["max_review"]:
        values_for_file.append("min")
        reviews_for_file.append(banks_n_reviews[bank]["min_review"])
    else:
        values_for_file.append("min")
        values_for_file.append("max")
        reviews_for_file.append(banks_n_reviews[bank]["min_review"])
        reviews_for_file.append(banks_n_reviews[bank]["max_review"])
df = pd.DataFrame({"value" : values_for_file, "review" : reviews_for_file})
df
```

```python
df.to_csv('selected_reviews.tsv', index=False)
```

```python
rewies_and_values = pd.read_csv("/Users/sofyakuznetsova/selected_reviews.tsv",
                   sep=",")
rewies_and_values
```

```python
import re
B = dict(rewies_and_values[rewies_and_values["value"] == "min"])
Min_words = []
for text in B['review']:
    text = text.lower()
    list_of_words = re.split(r'[^А-Яа-яёЁ]', text)
    to_delete = []
    for word in list_of_words:
        if word in STOPS or len(word) == 0:
            to_delete.append(word)
    for word in to_delete:
        list_of_words.pop(list_of_words.index(word))
    Min_words += list_of_words

A = dict(rewies_and_values[rewies_and_values["value"] == "max"])
Max_words = []
for text in A['review']:
    text = text.lower()
    list_of_words = re.split(r'[^А-Яа-яёЁ]', text)
    to_delete = []
    for word in list_of_words:
        if word in STOPS or len(word) == 0:
            to_delete.append(word)
    for word in to_delete:
        list_of_words.pop(list_of_words.index(word))
    Max_words += list_of_words
```
